```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/jbrownlee/Datasets/master/daily-min-temperatures.csv'
```

```python
data=pd.read_csv(url)
```

```python
data.head()
```

```python
# Assuming df is your DataFrame and 'date' is the column containing date strings
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.set_index('Date',inplace=True)
```

```python
data.head()
```

```python
data.plot(title='Daily Min Temperature',figsize=(8,6))
plt.xlabel("Time")
plt.ylabel("Temperature")
```

```python
series=pd.read_csv(url,header=0,index_col=0)
```

```python
series.head()
```

```python
plt.figure(figsize=(8,6))
plt.plot(series.index,series['Temp'],label='Temp')
plt.title("Daily Min Temperature")
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
fig,axes=plt.subplots(2,1,figsize=(8,6))
plot_acf(series['Temp'],lags=30,ax=axes[0],title="Autocorrelation Functiion(ACF)")
plot_pacf(series['Temp'],lags=30,ax=axes[1],title="Partial Autocorrelation Function(PACF)")
plt.tight_layout()
plt.show()
```
